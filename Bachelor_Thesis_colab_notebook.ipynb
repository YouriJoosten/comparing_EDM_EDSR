```python
from ultralytics import YOLO


model = YOLO("yolo11l-cls.pt")

# paths
data_dir = "/content/drive/MyDrive/BachelorThesisDrive/YOLO_training"
output_dir = "/content/drive/MyDrive/BachelorThesisDrive/YOLOv11_output"
# Training prompt:
results = model.train(
    data=data_dir,
    epochs=300,
    imgsz=32,
    batch=512,
    optimizer='AdamW',
    lr0=0.0001,
    save=True,
    save_period=10,
    device=0,
    amp=True,
    workers=16,
    project=output_dir,
    name='YOLOv11_classification_1',
    pretrained=True,
    plots=True,
    val=True,
    cache=True
)

print(results)
```
